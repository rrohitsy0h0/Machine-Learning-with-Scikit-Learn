```python
import pandas as pd #pandas was previously imported
#imported here incase this question is run first
```

(a) Read the .csv file as a Pandas dataframe.

```python
data=pd.read_csv('breast-cancer.csv')
data
```

(b) Convert all categorical attribute values to numerical values

```python
pd.set_option('future.no_silent_downcasting', True)
data2=data
data['diagnosis']=data['diagnosis'].replace('M','0')
data['diagnosis']=data['diagnosis'].replace('B','1')
data['diagnosis']=data['diagnosis'].astype(int)
data
# type(data['diagnosis'])
# print(data.dtypes)
```

c) Remove the column ID from the dataframe.

```python
data=data.drop(columns='id')
data
```

(d) Split the dataframe into training and test sets using random sampling and in the ration 2:1.

```python
from sklearn.model_selection import train_test_split
X=data.drop(columns='diagnosis')
y=data['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.333, random_state=42)

print("\nTrain and test data shapes:")
print("X_train:", X_train.shape, "X_test:", X_test.shape)

# X
```

(e) Normalize the training and test data using "Standard Scaling”

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
# Apply fit_transform
X_test_scaled = scaler.transform(X_test)
X_train_scaled_normalized=(X_train_scaled-X_train_scaled.min())/(X_train_scaled.max()-X_train_scaled.min())
X_test_scaled_normalized=(X_test_scaled-X_test_scaled.min())/(X_test_scaled.max()-X_test_scaled.min())

print("\nNormalized Train and test data using Standard scaling:")
print("X_train_scaled_normalized:\n", X_train_scaled_normalized, "\n\nX_test_scaled_normalized:\n",X_test_scaled_normalized)
```

(f) Build the classification models using SVM, LR, RF, DT, XGBoost, Gradient Boosting algorithm.

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score
from sklearn.metrics import roc_auc_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
svm_model = SVC()
svm_model.fit(X_train_scaled, y_train)
svm_predictions = svm_model.predict(X_test_scaled)
y_pred = svm_predictions
print("\nSVM Predictions (First 10):", y_pred[:10])

def plot_confusion_matrix(y_true, y_pred, title):
    conf_matrix = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(5, 4))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Prediction labels')
    plt.ylabel('True lables')
    plt.title(title)
    plt.show()
plot_confusion_matrix(y_test, y_pred, "SVM Confusion Matrix")


svm_accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy:", svm_accuracy)

svm_precision = precision_score(y_test, y_pred, average="weighted")
print("\nPrecision: ", svm_precision)
```

```python
plot_confusion_matrix(y_test, y_pred, "Logistic Regression Confusion Matrix")

lr_accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy:", lr_accuracy)
lr_precision = precision_score(y_test, y_pred, average="weighted")
print("\nPrecision: ", lr_precision)

cm = confusion_matrix(y_test, y_pred)
specificity = []
for i in range(len(cm)):
    tn = np.sum(cm) - np.sum(cm[i, :]) - np.sum(cm[:, i]) + cm[i, i]
    fp = np.sum(cm[:, i]) - cm[i, i]
    specificity.append(tn / (tn + fp))
    lr_specificity = np.mean(specificity)
print("\nSpecificity: ", lr_specificity)

print("\nLogistic Regression Classification Report:", classification_report)
print("Classification with Logistic Regression is done!")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_scaled, y_train)
rf_predictions = svm_model.predict(X_test_scaled)
y_pred = rf_predictions
print("\nRandom Forest Predictions (First 10):", y_pred[:10])
plot_confusion_matrix(y_test, y_pred, "Random Forest Confusion Matrix")

rf_accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy:", rf_accuracy)

rf_precision = precision_score(y_test, y_pred, average="weighted")
print("\nPrecision: ", rf_precision)

cm = confusion_matrix(y_test, y_pred)
specificity = []
for i in range(len(cm)):
    tn = np.sum(cm) - np.sum(cm[i, :]) - np.sum(cm[:, i]) + cm[i, i]
    fp = np.sum(cm[:, i]) - cm[i, i]
    specificity.append(tn / (tn + fp))
    rf_specificity = np.mean(specificity)
print("\nSpecificity: ", rf_specificity)

# Report the summary of all evaluation:
print("\nRandom Forest Classification Report:", classification_report)
print("Classification with Random Forest is done!")
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_scaled, y_train)
dt_predictions = dt_model.predict(X_test_scaled)
y_pred = dt_predictions

print("\nDecision Tree Predictions (First 10):", y_pred[:10])

plot_confusion_matrix(y_test, y_pred, "Decision Tree Confusion Matrix")

dt_accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy:", dt_accuracy)

dt_precision = precision_score(y_test, y_pred, average="weighted")
print("\nPrecision: ", dt_precision)


cm = confusion_matrix(y_test, y_pred)
specificity = []
for i in range(len(cm)):
    tn = np.sum(cm) - np.sum(cm[i, :]) - np.sum(cm[:, i]) + cm[i, i]
    fp = np.sum(cm[:, i]) - cm[i, i]
    specificity.append(tn / (tn + fp))
    dt_specificity = np.mean(specificity)
print("\nSpecificity: ", dt_specificity)

print("\nDecision Tree Classification Report:", classification_report)
print("Classification with Decision Tree is done!")
```
